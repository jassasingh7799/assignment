````python
#5)Explain a iterator function
# Answer-
# Iterator function is a python function in python which help us to iterator over a sequence of values one at a time.It is commonly done using __iter__() and __next__() method.
# 1) __iter__()
# It return the iterator object and it is required for an object to be an iterable.
# 2) __next__()
# It will return the next value from the sequence.If there are no more items in the sequence it will raises the stopiteration exception to signal the end of the iteration.

# Example-
class NumberIterator:
    def __init__(self,limit):
        self.limit=limit
        self.current=0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.limit:
            value = self.current
            self.current+=1
            return value
        else:
            raise StopIteration
num =NumberIterator(5)
for i in num:
    print(i)
````

#27) what is encapsulation in oops?
Ans-In Object-Oriented Programming (OOP), encapsulation is a fundamental concept that refers to the bundling of data (variables) and methods (functions) that operate on that data into a single unit or class. Additionally, encapsulation restricts direct access to some of the object's components, which can help prevent unintended interference and misuse of the data.

Encapsulation means:

Bundling data and methods together: A class in OOP holds both the data (attributes or properties) and the functions (methods) that operate on the data.

Hiding the internal state (data hiding): The internal state of an object is often hidden from the outside world by making the attributes private or protected, and only accessible through specific public methods (often called getters and setters).

Control access to the data: By controlling how the data is accessed and modified, encapsulation can enforce rules about what values can be assigned to the data, thereby maintaining the integrity of the object.

Private vs. Public: The internal data (attributes) of an object can be marked as private or protected, meaning it cannot be accessed directly from outside the class. Instead, the object provides public methods to interact with that data.

Getter and Setter Methods: These are public methods that allow controlled access to private variables. A getter method retrieves the value of an attribute, and a setter method modifies the value of an attribute, often with additional validation.


Benefits of Encapsulation:
Data Hiding: It helps protect an object's state by hiding the internal data and requiring controlled access through methods.

Modularity: The internal workings of an object are encapsulated, meaning you can change the implementation without affecting other parts of the program.

Improved Maintenance: By restricting direct access to data, encapsulation allows for easier debugging and maintenance of code. You can change the way data is stored internally without changing how it’s accessed by others.

Flexibility and Reusability: Encapsulation allows the creation of more flexible and reusable code by isolating the internal details and exposing only the necessary interface.

````python
#example of encapsulation 

class Person:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age

    def get_name(self):
        return self.__name

    def set_name(self,name):
        self.__name=name

    def get_age(self):
        return self.__age

    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print('Age cannot be negative :',age)

p=Person('Jasmeet',25)
print('initial name:',p.get_name())

p.set_name('Jasmeet Singh Arora')
print('Updated name:',p.get_name())
print('initial age:',p.get_age())

p.set_age(26)
print('Updated age:',p.get_age())
````

#28) Explain the use of access modifiers in Python classes?
Ans-
Access modifiers are used to control the visibility and accessibility of class attributes and methods. They help define whether certain parts of a class (such as attributes and methods) can be accessed from outside the class or not. While Python does not have explicit access modifiers like some other languages.e.g., public, private, protected like java and c++ it uses a system of naming conventions to indicate the intended visibility.

The three main levels of access control in Python are:
1) Public Access
2) Protected Access
3) Private Access

1) Public Access-
    1) Attributes/Methods: Any attribute or method in a Python class is public, meaning it can be accessed and modified directly from outside the class.
    2)Syntax: Public members are simply declared without any leading underscores(make)
    example of public-
        class Car:
            def __init__(self,make,model):
                self.make=make
                self.model=model
    3)Public access means that anyone who uses the class can access these attributes and methods directly.

2) Private Access:
    1) Attributes/Methods: To make an attribute or method private, Python uses double underscores (__) as a naming convention. This triggers name mangling, which makes the attribute or method name harder to access from outside the class.
    2) Syntax: __make
    example of private-
        class Car:
            def __init__(self,make,car)
                self.__make=make
                self.__car=car
    3) Private access makes it harder (but not impossible) to access or modify the attribute or method directly from outside the class. The __ prefix triggers name mangling, which internally changes the name of the attribute (e.g., self.__balance becomes self._BankAccount__balance), making it less likely to be accessed inadvertently.

3) Protected Access:
    1) Attributes/Methods: A convention in Python to indicate that an attribute or method is protected is by prefixing it with a single underscore (_). This is a hint to other programmers that these members are intended for internal use within the class or its subclasses, but it does not prevent access.
    2) Syntax: _attribute_name
    example of protected-
        class Car:
            def __init__(self,make,car):
                self._make=make
                self._car=car
    3) Protected access is essentially a convention in Python. It tells developers that they should treat the attribute as protected (meaning it’s intended for use within the class or subclasses).

We have the following why we use access specifier-
Encapsulation: Helps in encapsulating the data and restricting direct access, providing more control over how the data is accessed or modified. This is a key principle of Object-Oriented Programming.

Clear Intentions: Using access modifiers communicates to other developers how the class members should be used, even if Python doesn’t strictly enforce them. For instance, _attribute signals that a member is protected, while __attribute signals that it should be treated as private.

Code Safety: By restricting access to certain parts of the code, you can prevent unintended side effects, reduce the risk of bugs, and ensure that the internal state of an object is only modified in a controlled manner.

#29) What is inheritance in OOPs?

Ans-
Inheritance in Object-Oriented Programming (OOP) is a mechanism where one class (the subclass or derived class) can inherit properties and behaviors (i.e., fields and methods) from another class (the superclass or base class). It allows the creation of a new class based on an existing class, promoting code reuse and establishing a natural hierarchy.

Key Concepts of Inheritance:
1) Base Class (Parent Class or Superclass):
The class whose properties and methods are inherited by another class.
It defines common attributes and behaviors that can be shared with other classes.
    
2) Derived Class (Child Class or Subclass):
The class that inherits properties and behaviors from the base class.
It can also have additional properties and methods that are specific to it, or it can override the inherited methods to change or extend functionality.
    
3) Access to Inherited Members:
The subclass inherits both instance variables and methods from the base class.
It can access the inherited members directly unless they are marked with an access modifier (like private) that restricts access.

4)Overriding:
A subclass can override methods from the superclass to provide its own implementation. This is done by defining a method in the subclass with the same name and signature as in the base class.

5) Method overloading:
Method overloading in Python refers to defining a method with the same name but different argument types or counts. Python does not support traditional method overloading like in other languages. Instead, it achieves similar functionality using default arguments, variable-length arguments (`*args`), or manual argument checks.

6)Constructor Inheritance:
Constructors are not inherited directly, but the subclass can call the constructor of the superclass using special syntax (e.g., super() in Java or super() in Python).

Types of Inheritance:

Single Inheritance: A class inherits from one superclass.
Multiple Inheritance: A class inherits from more than one superclass (supported in languages like Python but not in Java).
Multilevel Inheritance: A class inherits from a class that is already a subclass of another class.
Hierarchical Inheritance: Multiple subclasses inherit from a single superclass.
Hybrid Inheritance: A combination of two or more types of inheritance.

#30) Define polymorphism in OOPs?
Ans-
Polymorphism in Object-Oriented Programming (OOP) is the ability of different classes to respond to the same method or function in different ways. It allows objects of different types to be treated as objects of a common superclass, with each subclass providing its own implementation of the method.

In simple terms, polymorphism means "many forms", where a single method or operation behaves differently based on the type of object that invokes it.

Types of Polymorphism:

Compile-Time Polymorphism (Static Polymorphism):
Resolved during compile time.
Method Overloading: Multiple methods with the same name but different parameter lists (number or type of parameters).
Operator Overloading: Changing the behavior of operators (e.g., + operator).
                                                                                       
Runtime Polymorphism (Dynamic Polymorphism):
Resolved at runtime.
Method Overriding: A subclass provides its own implementation of a method that is already defined in the superclass. The correct method is called based on the object type at runtime.

#31) Explain method overriding in Python?
Ans-
Method overriding in Python is a feature that allows a subclass to provide a specific implementation for a method that is already defined in its superclass. This is part of runtime polymorphism (also known as dynamic polymorphism), where the method of the subclass overrides the method in the superclass.

When a method in a superclass is overridden in a subclass, the subclass's version of the method is called, even if the object is of the superclass type. This allows subclasses to modify or extend the behavior of methods inherited from the superclass.

Points of Method Overriding:
The method in the subclass must have the same name, return type, and parameters as the method in the superclass.
When the method is called on an instance of the subclass, the subclass’s method is invoked (even if the reference is of the superclass type).
The super() function can be used to call the superclass's version of the method if needed.

Reason fro using method overriding-
Customization: Method overriding allows subclasses to provide their own implementation of a method inherited from the superclass, customizing it to fit the subclass's needs.
Polymorphism: It enables polymorphic behavior, where the same method name can produce different results depending on the object type, even if the reference is of the superclass type.

````python
#Example of method overriding

class Animal:
    def Sound(self):
        print('Animal makes sound')

class Dog(Animal):
    def Sound(self):
        print('Dog barks')

class Cat(Animal):
    def Sound(self):
        print('cat meows')

#creating object of Dog and cat class
d=Dog()
c=Cat()

#call the overriden methods
d.Sound()
c.Sound()
````

````python
#use of super in method overriding

class Animal:
    def Sound(self):
        print('Animal makes sound')

class Dog(Animal):
    def Sound(self):
        super().Sound() #calls the Sound method from Animal
        print('Dog barks')

class Cat(Animal):
    def Sound(self):
        print('cat meows')

#creating object of Dog and cat class
d=Dog()
c=Cat()

#call the overriden methods
d.Sound()
c.Sound()
````

#37) What is abstraction in Python? How is it implemented ?

Abstraction in Python refers to the concept of hiding complex implementation details and exposing only the necessary functionalities to the user. The goal is to provide a simplified interface, while keeping the internal workings of a system hidden. This helps in reducing complexity and improving the maintainability of the code.
abstraction is commonly implemented using abstract classes and abstract methods, which are part of the abc (Abstract Base Class) module.

Key Concepts:
Abstract Class: This is a class that cannot be instantiated on its own and is designed to be inherited by other classes. It may contain one or more abstract methods, which are methods that are declared but not implemented in the abstract class.
Abstract Method: These are methods that are declared in an abstract class, but do not have any implementation. Any class that inherits from the abstract class is required to provide an implementation for these abstract methods.

Reason of why to use Abstraction:
Separation of Concerns: Abstraction allows you to separate the "what" from the "how." For example, you can define the concept of "Shape" and leave the specific details (like how to calculate the area) to the subclasses.

Code Reusability and Maintenance: When the interface remains consistent, you can reuse the same code while updating the underlying implementation without affecting other parts of the code.

Encapsulation: Abstraction helps in encapsulating implementation details, preventing users from needing to understand complex code, which makes the system easier to maintain and extend.

Encapsulation helps in protecting the internal state of an object and controlling how the data is accessed and modified, which makes your code more maintainable, secure, and modular. 

-Achieving Encapsulation in Python

1). Public Attributes and Methods
   By default, all attributes and methods in Python are public, meaning they can be accessed and modified directly from outside the class.

Example-
class Person:
    def __init__(self, name, age):
        self.name = name  # Public attribute
        self.age = age    # Public attribute

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Usage
p = Person("Alice", 30)
print(p.name)   # Accessing public attribute
p.greet()       # Calling public method

2). Private Attributes and Methods (Name Mangling)
   To prevent direct access to attributes and methods, Python uses name mangling. By prefixing an attribute or method with double underscores (`__`), Python changes the name of the attribute or method internally to make it less accessible from outside the class.

   While it is still possible to access private members using name mangling (i.e., using the mangled name), it is not recommended as it breaks encapsulation.

class Person:
    def __init__(self, name, age):
        self.__name = name  # Private attribute (name mangling)
        self.__age = age    # Private attribute (name mangling)

    def __increment_age(self):  # Private method (name mangling)
        self.__age += 1

    def greet(self):
        print(f"Hello, my name is {self.__name} and I am {self.__age} years old.")

    def have_birthday(self):
        self.__increment_age()  # Encapsulating internal logic
        print(f"Happy birthday, {self.__name}! You are now {self.__age} years old.")

p=Person('Jasmeet Singh Arora',25)
p.greet()               # Works fine
p.have_birthday()       # Internal logic is encapsulated


3) Protected Attributes and Methods (Single Underscore)
   Python uses a single underscore (`_`) as a convention to indicate that an attribute or method is protected — meaning it is intended to be used within the class or by subclasses, but it is still technically accessible from outside.

class Person:
    def __init__(self, name, age):
        self._name = name  # Protected attribute (by convention)
        self._age = age    # Protected attribute (by convention)

    def _increment_age(self):  # Protected method (by convention)
        self._age += 1

    def greet(self):
        print(f"Hello, my name is {self._name} and I am {self._age} years old.")

# Usage

4) Using Properties for Getter and Setter Methods
   To control access to private attributes, you can use **properties**. Properties allow you to define methods that can be accessed like attributes, while still allowing you to control the logic for getting or setting the values. This is a more Pythonic way to enforce encapsulation and provide controlled access to object attributes.

   You can use the `property()` function or the `@property` decorator to create getter and setter methods.

Example-
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def name(self):  # Getter for 'name'
        return self.__name

    @name.setter
    def name(self, value):  # Setter for 'name'
        if not value:
            raise ValueError("Name cannot be empty")
        self.__name = value

    @property
    def age(self):  # Getter for 'age'
        return self.__age

    @age.setter
    def age(self, value):  # Setter for 'age'

            raise ValueError("Age cannot be negative")
        self.__age = value

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Usage
p = Person("Jasmeet Singh Arora", 25)
p.greet()  # Outputs: Hello, my name is Jasmeet Singh Arora and I am 25 years old.

# You can now use 'name' and 'age' like attributes, but behind the scenes, they're being managed via getter and setter methods
p.name = "Jasmeet Singh"    # Setter for 'name' called
p.age = 26        # Setter for 'age' called

#48) Can encapsulation be bypassed in Python? If so, how?

Ans-
Yes, encapsulation can technically be bypassed in Python, but it is important to understand that Python’s approach to encapsulation is convention-based rather than enforced by the language. In other words, Python relies on naming conventions (such as the use of underscores) and mechanisms like name mangling to signal the intended level of access to class attributes or methods, but these mechanisms can be circumvented.

Here are the main ways encapsulation can be bypassed in Python:

1) Accessing Private Attributes (Name Mangling)
   Python uses name mangling for private attributes and methods. If an attribute or method is prefixed with `__` (double underscores), Python internally renames it by adding the class name as a prefix. This is done to avoid accidental name clashes in subclasses.

Example-
   class Person:
       def __init__(self, name, age):
           self.__name = name  # Private attribute
           self.__age = age    # Private attribute

       def greet(self):
           print(f"Hello, my name is {self.__name} and I am {self.__age} years old.")

   Python mangles the names of the private attributes to make them harder to access directly:


   p = Person("Jasmeet Singh Arora", 25)
   # print(p.__name)   # AttributeError: 'Person' object has no attribute '__name'

   # However, we can access the mangled names directly:
   print(p._Person__name)  # Outputs: Jasmeet Singh Arora
   print(p._Person__age)   # Outputs: 25

   - Even though the `__name` and `__age` attributes are private, they can still be accessed directly via their mangled names (`_Person__name` and `_Person__age`).
   - This bypasses encapsulation because the private members are still accessible if you know their mangled names, even though this is generally considered bad practice.

2) Accessing Protected Attributes
   In Python, protected attributes are defined with a single underscore (`_`). This is purely a convention that signals that these attributes should not be accessed directly from outside the class. However, unlike private attributes, protected attributes can still be accessed without restriction.

Example
   class Animal:
       def __init__(self, name):
           self._name = name  # Protected attribute

   a = Animal("Lion")
   print(a._name)  # Outputs: Lion (no error)
   ```

   - The `_name` attribute is considered protected, meaning it is intended to be used by the class and its subclasses. 
   - However, since there is no enforcement of this convention in Python, other parts of the program (even outside the class) can access and modify protected attributes.

3) Using `__dict__` for Direct Access
   Every Python object has a special attribute called `__dict__`, which contains a dictionary of all the objects attributes and methods. This provides another way to bypass encapsulation and access both private and protected attributes directly.

Example-
   class Person:
       def __init__(self, name, age):
           self.__name = name  # Private attribute

   print(p.__dict__)  # {'_Person__name': 'Jasmeet Singh Arora', '_Person__age': 25}

   - The `__dict__` attribute allows access to all the attributes of an object, including private ones. 
   - In this example, even though the `__name` and `__age` attributes are private, they can still be accessed directly via `p.__dict__`.

4) Using `getattr()` and `setattr()` Functions
   Python provides built-in functions like `getattr()` and `setattr()` that can be used to access and modify an objects attributes dynamically. These functions can be used to bypass encapsulation as well.

Example-
   class Person:
       def __init__(self, name, age):

   p = Person("Jasmeet Singh Arora", 25)

#52) Create a decorator in Python that adds functionality to a simple function by printing a message before and after the function execution ?

Ans-
A decorator in Python is a function that allows you to modify or extend the behavior of another function without permanently modifying it. It essentially wraps the function, adding extra functionality before or after its execution.
To create a decorator that prints a message before and after the execution of a function, you can define a function that takes another function as its argument, and then returns a new function that wraps the original one.

````python
#Example1

# Define the decorator
def print_message_decorator(func):
    def wrapper(*args, **kwargs):
        # Print message before function execution
        print("Before executing the function.")
        
        # Call the original function
        result = func(*args, **kwargs)
        
        # Print message after function execution
        print("After executing the function.")
        
        return result
    return wrapper

# Apply the decorator to a simple function
@print_message_decorator
def say_hello(name):
    print(f"Hello, {name}!")

# Call the decorated function
say_hello("Jasmeet Singh Arora")
````

#71) How does encapsulation relate to property decorators in Python? Provide an example showcasing encapsulation using property decorators?
Ans-
Encapsulation is one of the core principles of object-oriented programming (OOP), and it refers to the concept of bundling data (attributes) and the methods that operate on the data into a single unit (a class), and restricting direct access to some of the classs components. This is done to protect the integrity of the objects data and prevent outside code from directly modifying the internal state in unexpected or invalid ways.
In Python, encapsulation is often achieved through private and protected attributes (indicated by a leading underscore `_`), and more commonly through getter, setter, and deleter methods (often decorated with `@property`, `@<property>.setter`, and `@<property>.deleter`).

Property Decorators and Encapsulation-

The `@property` decorator in Python allows for **encapsulation** by enabling you to manage how an attribute is accessed and modified. Using property decorators, you can:

1. Restrict access to an attribute by providing a controlled getter or setter method.
2. Add logic or validation to the getter/setter methods to ensure that the data is always valid.
3. Keep the internal representation of an attribute hidden, so that external code can only interact with the object through its public interface (i.e., getter/setter methods).

Example of Encapsulation Using Property Decorators-
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self._balance = balance  # The internal variable for storing balance

    @property
    def balance(self):
        """Getter for the balance."""
        return self._balance

    @balance.setter
    def balance(self, amount):
        """Setter for the balance with validation."""
        if amount < 0:
            raise ValueError("Balance cannot be negative!")
        self._balance = amount

    def deposit(self, amount):
        """Method to deposit money into the account."""
        if amount <= 0:
            raise ValueError("Deposit amount must be positive!")
        self.balance += amount  # Using the setter method to update the balance

    def withdraw(self, amount):
        """Method to withdraw money from the account."""
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive!")
        if amount > self.balance:
            raise ValueError("Insufficient balance!")
        self.balance -= amount  # Using the setter method to update the balance
